```python
# Converter as colunas de data para o formato datetime
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
df['order_approved_at'] = pd.to_datetime(df['order_approved_at'])
df['order_delivered_carrier_date'] = pd.to_datetime(df['order_delivered_carrier_date'])
df['order_delivered_customer_date'] = pd.to_datetime(df['order_delivered_customer_date'])
df['order_estimated_delivery_date'] = pd.to_datetime(df['order_estimated_delivery_date'])

# Separar as informações com base nos espaços
df[['data_compra', 'hora_compra']] = df['order_purchase_timestamp'].astype(str).str.split(expand=True)
df[['data_aprovacao', 'hora_aprovacao']] = df['order_approved_at'].astype(str).str.split(expand=True)
df[['data_entrega_transportadora', 'hora_entrega_transportadora']] = df['order_delivered_carrier_date'].astype(str).str.split(expand=True)
df[['data_entrega_cliente', 'hora_entrega_cliente']] = df['order_delivered_customer_date'].astype(str).str.split(expand=True)
df[['data_estimada_entrega', 'hora_estimada_entrega']] = df['order_estimated_delivery_date'].astype(str).str.split(expand=True)

# Excluir as colunas originais de data e hora
df = df.drop(['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date'], axis=1)
```
