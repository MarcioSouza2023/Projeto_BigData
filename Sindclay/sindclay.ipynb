```python
# Exemplo de DataFrame
data = {'geolocation_city': ['sao paulo', 'sãopaulo', 'sa£o paulo', 'río de janeiro']}
df = pd.DataFrame(data)

# Função para remover caracteres especiais e acentuação
def remover_acentos(texto):
    return unidecode(texto)

# Aplicar a função à coluna 'geolocation_city'
df['geolocation_city'] = df['geolocation_city'].apply(remover_acentos)

# Exibir o DataFrame resultante
print(df)
```

```python
# Selecione a coluna desejada
unique_cities = df['geolocation_city'].unique()

# Crie um DataFrame a partir do array
df_unique_cities = pd.DataFrame({'Unique Cities': unique_cities})

# Exiba o DataFrame
df_unique_cities
```
